**Here is the data dictionary for the given columns**:

`Month`: This column represents the month of the year during which the data was recorded. 

`Year`: The "Year" column denotes the calendar year corresponding to the data recorded in the dataset.

`Total Orders`: This column refers to the total number of orders placed during the specified month and year. It is a count of the individual transactions processed by the business within the given time frame.

`Gross Sales`: The column indicates the total revenue generated from the sales transactions within the specified month and year, before accounting for any discounts, returns, or shipping charges.

`Discounts`: This column represents the total amount of discounts applied to the sales transactions during the specified month and year. Discounts are deductions from the gross sales amount and are typically provided to customers as incentives or promotions.

`Returns`: The "Returns" column denotes the total value of returned merchandise or products during the specified month and year. Returns occur when customers send back purchased items for various reasons, such as defects, dissatisfaction, or change of mind.

`Net Sales`: Net Sales refers to the total revenue generated from sales transactions after accounting for discounts and returns. It represents the actual amount of revenue earned by the business from selling its products or services during the specified period.

`Shipping`: This column indicates the total shipping charges incurred by the business for fulfilling orders during the specified month and year. Shipping charges typically include costs associated with packaging, handling, and delivery of the ordered items to customers.

`Total Sales`: The "Total Sales" column represents the overall sum of gross sales, discounts, returns, and shipping charges, providing a comprehensive view of the total revenue generated by the business during the specified month and year.

```python
import matplotlib.pyplot as plt
import seaborn as sb
import warnings

warnings.filterwarnings("ignore")

%matplotlib inline
plt.figure(figsize=(14,7))
```

```python
num = sales_df.select_dtypes(include=['float64'])
num_columns = [i for i in num.columns]
```

```python
num_columns
```

```python
# Histogram for numerical columns

for column in num_columns:
    plt.hist(sales_df[column], bins=10, color='grey', edgecolor='black', alpha=0.7)
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

```python
# Density plot for numerical columns

for column in num_columns:
    sb.kdeplot(sales_df[column],color='grey')
    plt.title(f'Density plot of {column}')
    plt.show()
```

```python
# Box plot for numerical columns

for column in num_columns:
    plt.figure(figsize=(20,8))
    plt.boxplot(sales_df[column],notch=True, patch_artist=True, boxprops=dict(facecolor='grey', color='grey'), medianprops=dict(color='red', linewidth=2), whiskerprops=dict(color='green', linewidth=1.5, linestyle='--'), capprops=dict(color='blue', linewidth='1.5'), flierprops=dict(marker='o',markersize=8, markerfacecolor='yellow', markeredgecolor='black'), vert=False)
    plt.title(f'Boxplot of {column}')
    plt.ylabel(column)
    plt.grid(True, linestyle='--', linewidth=0.5)
    plt.tight_layout()
    plt.show()
```
